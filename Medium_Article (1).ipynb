```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style='whitegrid')

# Try to import the Colab file upload widget; if it fails, skip it.
try:
    from google.colab import files
    print("Google Colab environment detected. You can upload the file using the widget.")
    # Uncomment the next two lines if you need to upload the file interactively
    # uploaded = files.upload()
    # filename = list(uploaded.keys())[0]
    filename = 'onlinedeliverydata.csv'
except ImportError:
    print("Google Colab not detected. Assuming the file is in the current directory.")
    filename = 'onlinedeliverydata.csv'

# Load the dataset
df = pd.read_csv(filename, encoding='ascii')

# Display the first few rows of the dataframe
print("Head of the dataframe:")
print(df.head())

# Display summary information about the dataframe
print("\nDataframe Info:")
print(df.info())

# Extract numeric columns for potential visualizations
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
print("\nNumeric columns:", list(numeric_cols))

# Visualization 1: Pairplot of numeric columns (if more than one column exists)
if len(numeric_cols) >= 2:
    sns.pairplot(df[numeric_cols].dropna())
    plt.show()

# Visualization 2: Distribution of 'Age' (if exists)
if 'Age' in df.columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(df['Age'].dropna(), kde=True, bins=20)
    plt.title("Distribution of Age")
    plt.xlabel("Age")
    plt.ylabel("Frequency")
    plt.show()

# Visualization 3: Count plot for 'Gender' (if exists)
if 'Gender' in df.columns:
    plt.figure(figsize=(6, 4))
    sns.countplot(data=df, x='Gender')
    plt.title("Count of Genders")
    plt.xlabel("Gender")
    plt.ylabel("Count")
    plt.show()

# Visualization 4: Educational Qualifications count plot (if exists)
if 'Educational Qualifications' in df.columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, y='Educational Qualifications', order=df['Educational Qualifications'].value_counts().index)
    plt.title("Educational Qualifications Distribution")
    plt.xlabel("Count")
    plt.ylabel("Educational Qualifications")
    plt.show()

# Visualization 5: Scatter plot (latitude vs longitude), colored by Gender (if columns exist)
if 'latitude' in df.columns and 'longitude' in df.columns:
    plt.figure(figsize=(8, 6))
    if 'Gender' in df.columns:
        sns.scatterplot(data=df, x='longitude', y='latitude', hue='Gender', alpha=0.7)
    else:
        sns.scatterplot(data=df, x='longitude', y='latitude', alpha=0.7)
    plt.title("Geographical Distribution (Latitude vs Longitude)")
    plt.xlabel("Longitude")
    plt.ylabel("Latitude")
    plt.show()

print("Visualizations generated successfully.")
```

```python
# Code to visualize the customer's order time distribution using the 'Order Time' column from the dataset

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset; assuming file 'onlinedeliverydata.csv' is in the directory
# If already loaded, you can comment this out

df = pd.read_csv('onlinedeliverydata.csv', encoding='ascii')

# Check if 'Order Time' column exists in the dataframe
if 'Order Time' in df.columns:
    plt.figure(figsize=(10, 6))
    # Create a countplot for the 'Order Time'
    order_time_order = df['Order Time'].value_counts().index
    sns.countplot(data=df, x='Order Time', order=order_time_order, palette='coolwarm')
    plt.title('Preferred Order Time of Customers')
    plt.xlabel('Order Time')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
    print('Order Time visualization generated.')
else:
    print('Column "Order Time" not found in the dataset.')
```

```python
# Code to visualize the influence of discounts on ordering discussions and discount importance level

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset; assuming file 'onlinedeliverydata.csv' is in the directory

df = pd.read_csv('onlinedeliverydata.csv', encoding='ascii')

# We expect that the questions on discount have columns like 'More Offers and Discount' (which might indicate influence on ordering)
# and possibly a column for discount importance level, e.g. 'Discount Importance Level'.

# We'll check for those columns and plot accordingly

columns_to_plot = []
plot_titles = []

if 'More Offers and Discount' in df.columns:
    columns_to_plot.append('More Offers and Discount')
    plot_titles.append('Influence of Discounts on Ordering Discussions')
else:
    print('Column "More Offers and Discount" not found in the dataset.')

if 'Discount Importance Level' in df.columns:
    columns_to_plot.append('Discount Importance Level')
    plot_titles.append('Discount Importance Level')
else:
    print('Column "Discount Importance Level" not found in the dataset.')

# If no discount related columns, stop execution
if not columns_to_plot:
    print('No discount related columns found in the dataset to plot.')
else:
    n_plots = len(columns_to_plot)
    plt.figure(figsize=(8 * n_plots, 6))

    for idx, col in enumerate(columns_to_plot, start=1):
        plt.subplot(1, n_plots, idx)
        order_vals = df[col].value_counts().index
        sns.countplot(data=df, x=col, order=order_vals, palette='Set2')
        plt.title(plot_titles[idx-1])
        plt.xlabel(col)
        plt.ylabel('Count')
        plt.xticks(rotation=45)

    plt.tight_layout()
    plt.show()
    print('Discount influence visualizations generated.')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path = "/mnt/data/onlinedeliverydata.csv"  # Ensure correct file path
df = pd.read_csv('/content/onlinedeliverydata.csv')

# Inspect dataset structure
print("Columns in dataset:", df.columns)

# Select only numerical columns for correlation analysis
numeric_df = df.select_dtypes(include=['number'])

# Compute the correlation matrix
corr_matrix = numeric_df.corr()

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)

# Titles and labels
plt.title("Correlation Heatmap of Key Factors Influencing Orders")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Example true labels and predicted labels
y_true = ["No", "No", "Yes", "Yes", "Yes", "Yes", "No", "Yes", "No", "Yes"] * 8  # Adjust as needed
y_pred = ["No", "Yes", "Yes", "Yes", "Yes", "Yes", "Yes", "Yes", "Yes", "Yes"] * 8  # Adjust as needed

# Compute confusion matrix
cm = confusion_matrix(y_true, y_pred, labels=["No", "Yes"])

# Convert to DataFrame for visualization
cm_df = pd.DataFrame(cm, index=["No", "Yes"], columns=["No", "Yes"])

# Plot Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm_df, annot=True, fmt="d", cmap="Blues")

# Labels and title
plt.title("Confusion Matrix - Model Performance Evaluation")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")

# Show the plot
plt.show()
```
